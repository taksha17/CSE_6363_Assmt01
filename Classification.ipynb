```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
data = load_iris()
target = data.target
```

```python
plt.hist(target ,color='brown', orientation='horizontal')
plt.xlabel('txt6312_Class_on_the_X_Axis')
plt.ylabel('txt6312_Frequency_on_the_X_Axis')
plt.title('txt6312_Histogram_Chart_Classification_For_Target_Variable')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.1, random_state=42)
```

```python
# Linear Discriminant Analysis
components = 2
lda = LinearDiscriminantAnalysis(n_components=components)
lda.fit(X_train, y_train)
lda_predictions = lda.predict(X_test)
lda_acc = mean_squared_error(lda_predictions, y_test)

# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_predictions = nb.predict(X_test)

# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_predictions = logreg.predict(X_test)

# Mean Squared Error
lda_acc = mean_squared_error(lda_predictions, y_test)
nb_acc = mean_squared_error(nb_predictions, y_test)
logreg_acc = mean_squared_error(logreg_predictions, y_test)

print("Linear Discriminant Analysis Accuracy:", lda_acc)
print("Naive Bayes Accuracy:", nb_acc)
print("Logistic Regression Accuracy:", logreg_acc)
```

```python
lda = LinearDiscriminantAnalysis(solver='svd')
lda.fit(X_train, y_train)
```

```python
predicted = lda.predict(X_test)

# plt.scatter(range(len(predicted)), predicted, c=y_test)
colors = ['brown', 'green', 'red']
plt.scatter(range(len(predicted)), predicted, c=[colors[label] for label in y_test])
plt.xlabel('txt6312_Index_On_X_Axis')
plt.ylabel('txt6312_Predicted_Class_On_Y-Axis')
plt.title('txt6312_Scatter_Plot_Graph')
plt.show()



```
